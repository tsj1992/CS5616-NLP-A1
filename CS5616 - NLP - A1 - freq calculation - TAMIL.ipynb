### Reading the `tamil.txt` file

```python
datafile = open('tamil.txt', 'r', encoding = 'utf-8')
raw_data = datafile.read()
datafile.close()
```

### Removing all numbers

```python
import re
nonum_data = re.sub(r'\d+', '', raw_data)
```

### Removing all punctuation marks

```python
import string
nopunc_data = nonum_data.translate(str.maketrans("","", string.punctuation))
```

### Removing leading and trailing whitespaces

```python
stripped_data = nopunc_data.strip()
```

### NLTK Tokenizer

```python
import nltk
```

```python
words = nltk.word_tokenize(stripped_data)
```

```python
words
```

### Finding the frequencies of the words

```python
frequency = {}
```

```python
for word in words:
    count = frequency.get(word, 0)
    frequency[word] = count + 1
```

```python
for key, value in reversed(sorted(frequency.items(), key = lambda x: x[1])):
    print(key, value)
```

### Finding the frequencies of frequencies

```python
freq_of_freq = {}
```

```python
for item in frequency.items():
    for i in range(1, 11):
        if item[1] == i:
            count = freq_of_freq.get(f'{i}', 0)
            freq_of_freq[f'{i}'] = count + 1
    if 10 < item[1] and item[1] < 50:
        count = freq_of_freq.get('11-50', 0)
        freq_of_freq['11-50'] = count + 1
    if 51 < item[1] and item[1] < 100:
        count = freq_of_freq.get('51-100', 0)
        freq_of_freq['51-100'] = count + 1
    if 100 < item[1]:
        count = freq_of_freq.get('>100', 0)
        freq_of_freq['>100'] = count + 1
        
```

```python
freq_of_freq
```

### Finding rank

```python
sorted_frequency = {}
for key, value in reversed(sorted(frequency.items(), key=lambda x: x[1])):
    sorted_frequency[key] = value
```

```python
sorted_frequency
```

```python
wfr = {}
count = 1
for key, value in sorted_frequency.items():
    wfr[key] = (value, count)
    if count < 9:
        count += 1
    else:
        count = 1    
```

```python
wfr
```

```python
rank_order = 0
for w, (f, r) in wfr.items():
    wfr[w] = (f, r * (10**rank_order))
    if r % 10 == 9:
        rank_order += 1
```

```python
wfr
```

```python
freqs = []
ranks = []
for w, (f, r) in wfr.items():
    freqs.append(f)
    ranks.append(r)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.clf()
plt.xlabel('rank')
plt.ylabel('frequency')
plt.grid(True)
plt.title('Rank vs. Frequency for tamil.txt')
plt.plot(ranks[:20], freqs[:20])
plt.savefig('RvsF - Tamil.png', dpi=720)
```

```python
plt.clf()
plt.xlabel('rank (log scaled)')
plt.ylabel('frequency (log scaled)')
plt.yscale('log')
plt.xscale('log')
plt.grid(True)
plt.title('Rank vs. Frequency for tamil.txt (Log Scaled)')
plt.plot(ranks[:1000], freqs[:1000])
plt.savefig('RvsF (log scaled) - Tamil.png', dpi=720)
```
